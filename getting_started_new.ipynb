### Scraping data from Skytrax

If you visit [https://www.airlinequality.com] you can see that there is a lot of data there. For this task, we are only interested in reviews related to British Airways and the Airline itself.

If you navigate to this link: [https://www.airlinequality.com/airline-reviews/british-airways] you will see this data. Now, we can use `Python` and `BeautifulSoup` to collect all the links to the reviews and then to collect the text data on each of the individual review links.

```python
import requests #A library for making HTTP requests.
from bs4 import BeautifulSoup #A library for parsing HTML and XML documents.
import pandas as pd
import certifi
import os
```

```python
base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 10 #Number of pages you want to scrape.
page_size = 100 #Number of reviews per page.

reviews = []  #An empty list to store the text of the reviews.

# for i in range(1, pages + 1):
for i in range(1, pages + 1): #The loop runs from 1 to the number of pages specified (pages).

    print(f"Scraping page {i}") #i is the page number being scraped.

    # Create URL to collect links from paginated data
    url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}" #For each iteration, it constructs the URL for the specific page to scrape.
       #i is the current page number in the loop.
       #For example, if i is 1, it becomes /page/1/.
       #?sortby=post_date%3ADesc: A query parameter that specifies sorting the reviews by the post date in descending order (%3A is the URL-encoded representation of the colon :).
       #&pagesize={page_size}: A query parameter that sets the number of reviews per page. page_size is the variable specifying this number (in this case, 100).
       #eg : https://www.airlinequality.com/airline-reviews/british-airways/page/1/?sortby=post_date%3ADesc&pagesize=100

    # Collect HTML data from this page
    response = requests.get(url, verify=False)  # Makes a GET request to fetch the HTML content of the page.

    # Parse content
    content = response.content #Stores the HTML content of the page.
    parsed_content = BeautifulSoup(content, 'html.parser') #Parses the HTML content using BeautifulSoup.
    for para in parsed_content.find_all("div", {"class": "text_content"}):
        reviews.append(para.get_text())
    
    print(f"   ---> {len(reviews)} total reviews")
    
#parsed_content.find_all("div", {"class": "text_content"}): Finds all the div elements with the class text_content which contains the review texts.
#para.get_text(): Extracts the text content from each found div.
#reviews.append(para.get_text()): Appends the extracted text to the reviews list.
#print(f" ---> {len(reviews)} total reviews"): Prints the total number of reviews collected so far.
```

```python
df = pd.DataFrame()  #Creates an empty DataFrame.
df["reviews"] = reviews  #Adds the collected reviews to the DataFrame under the column reviews.
df.head() # Displays the first few rows of the DataFrame (used for verification, not necessary for saving the file).
```

Data Cleaning

```python
import re #re: For regular expression operations to clean text.
```

```python
file_path = 'C:/Users/99906068/OneDrive - L&T Construction/Desktop/British Airways/BA_reviews.csv'
df = pd.read_csv(file_path)
```

```python
def clean_reviews(reviews):# Function to clean a single review
    reviews = re.sub(r'✅ Trip Verified ','', reviews)  # Remove "✅ Trip Verified" and any other unnecessary text
    reviews = re.sub(r'\|\|', '', reviews)
    reviews = reviews.strip()
    return reviews
```

```python
if 'reviews' in df.columns:
    # Apply the cleaning function to the reviews column
    df['reviews'] = df['reviews'].apply(clean_reviews)
    print(df.head())
```

```python
from textblob import TextBlob

# Function to analyze sentiment
def get_sentiment(reviews):
    analysis = TextBlob(reviews)
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity < 0:
        return 'negative'
    else:
        return 'neutral'

# Apply sentiment analysis to the reviews
df['sentiment'] = df['reviews'].apply(get_sentiment)

# Sentiment distribution
sentiment_counts = df['sentiment'].value_counts()
print(sentiment_counts)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Vectorize the text data
vectorizer = CountVectorizer(stop_words='english', max_features=1000)
text_data = vectorizer.fit_transform(df['reviews'])

# Fit LDA model
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(text_data)

# Display the top words in each topic
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

no_top_words = 10
display_topics(lda, vectorizer.get_feature_names_out(), no_top_words)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Combine all reviews into a single string
all_reviews = " ".join(review for review in df['reviews'])

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_reviews)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Reviews')
plt.show()
```
